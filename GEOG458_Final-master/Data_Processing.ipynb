```python
import geopandas as gpd
import pandas as pd
```

```python
shape = gpd.read_file("C:\Users\Kevin\Downloads\GEOG_458_Final\cb_2017_us_state_500k\cb_2017_us_state_500k.shp")
excel = pd.read_csv("C:\Users\Kevin\Downloads\GEOG_458_Final\exstall.csv", skiprows = 4)
rexel = pd.read_csv("C:\Users\Kevin\Downloads\GEOG_458_Final\exstall.csv", skiprows = 4)
```

```python
# Rename the statename column for easier merge later & subset the data to only include the top commodity exports of states &
# the states where the top commodity export is civilian aircrafts 

excel.rename(columns={'statename':'NAME'}, inplace = True)
clean = excel.loc[(excel['rank'] == 1) & (excel['hs6'] == 880000)]
```

```python
# Inserted columns with appropriate values to calculate the location quotient of each state for civilian aircraft exports

clean.insert(13,"State_Total_Export17", clean['val2017']/clean['share17']*100)
clean.insert(14,"US_Share_Aircraft17",120987.7)
clean.insert(15,"US_Total_Export17",1546272.96)
clean.insert(16,"LQ", (clean['share17']/100)/(clean['US_Share_Aircraft17']/clean['US_Total_Export17']))
```

```python
joinkey = 'NAME'
shape2 = shape.merge(clean, on=joinkey)
```

```python
# Subset the data to only include the top commodity exported for each state & inserted a column for the total export value
# of each state

rexel.rename(columns={'statename':'NAME'}, inplace = True)
clean1 = excel.loc[excel['rank'] == 1]
clean1.insert(13,"State_Total_Export17", clean1['val2017']/clean1['share17']*100)

shape3 = shape.merge(clean1, on=joinkey)
```

```python
shape2.to_file("C:\\Users\\Kevin\\Downloads\\GEOG_458_Final\\Static.shp")
shape3.to_file("C:\\Users\\Kevin\\Downloads\\GEOG_458_Final\\Interactive.shp")
```
